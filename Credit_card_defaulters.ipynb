```python
# importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# creation of dataframe from csv file
df_creditcard = pd.read_csv('UCI-credit-card.csv')
df_creditcard.head()
```

```python
# checking datatypes and presence of null-values
df_creditcard.info()
```

```python
df_creditcard.drop('ID', axis = 1)
```

```python
#summary statistics of the data
df_creditcard.describe()
```

```python
# Number of payment defaulters
df_creditcard['default.payment.next.month'].value_counts()
```

```python
sns.countplot(x="default.payment.next.month", data=df_creditcard)
```

```python
df_creditcard['default.payment.next.month'].value_counts().plot(kind='pie', labels=['',''], autopct='%1.1f%%')
plt.legend(labels=['No default', 'default'])
plt.show()
```

```python
df_creditcard.groupby('SEX')['default.payment.next.month'].value_counts(normalize = True)
```

```python
#% of payment defaults by gender
sns.barplot(x='SEX', y='default.payment.next.month', data = df_creditcard, ci=None)
plt.xticks([0,1],["Male", "Female"])
```

```python
df_creditcard.groupby('EDUCATION')['default.payment.next.month'].value_counts(normalize = True)
```

```python
education = ['Unknown', 'Grad School', 'University', 'High School', 'Others', 'UNKNOWN', '']
sns.barplot(x='EDUCATION', y='default.payment.next.month', data = df_creditcard, label=education, ci=None)
plt.xticks([0,1,2,3,4,5,6], education, rotation=45)
```

```python
df_creditcard.groupby('AGE')['default.payment.next.month'].value_counts(normalize = True)
```

```python
sns.kdeplot(df_creditcard.loc[df_creditcard['default.payment.next.month']==0, 'AGE'], label='No Default', shade=True)
sns.kdeplot(df_creditcard.loc[df_creditcard['default.payment.next.month']==1, 'AGE'],label='Default', shade=True)
plt.legend()
```

```python
fig, ax = plt.subplots(3,2,figsize=(15,10))


sns.scatterplot(x='PAY_AMT1',y='BILL_AMT1', data=df_creditcard, ax=ax[0,0], hue='default.payment.next.month')

sns.scatterplot(x='PAY_AMT2',y='BILL_AMT2', data=df_creditcard, ax=ax[0,1], hue='default.payment.next.month')

sns.scatterplot(x='PAY_AMT3',y='BILL_AMT3', data=df_creditcard, ax=ax[1,0], hue='default.payment.next.month')

sns.scatterplot(x='PAY_AMT4',y='BILL_AMT4', data=df_creditcard, ax=ax[1,1], hue='default.payment.next.month')

sns.scatterplot(x='PAY_AMT5',y='BILL_AMT5', data=df_creditcard, ax=ax[2,0], hue='default.payment.next.month')

sns.scatterplot(x='PAY_AMT6',y='BILL_AMT6', data=df_creditcard, ax=ax[2,1], hue='default.payment.next.month')
```
